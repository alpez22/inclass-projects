```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

def generate_erdos_renyi_self_loops(n, p, q):
    G = nx.Graph()
    G.add_nodes_from(range(n))
    
    #add edges between distinct nodes with probability p
    for i in range(n):
        for j in range(i + 1, n):
            if np.random.rand() < p:
                G.add_edge(i, j)
    
    #add self-loops with probability q
    for i in range(n):
        if np.random.rand() < q:
            G.add_edge(i, i)
    
    return G
```

```python
n = 100  #num of nodes
p = 0.2  #prob of edge between distinct nodes
q = 0.1  #prob of self-loop

G = generate_erdos_renyi_self_loops(n, p, q)

plt.figure(figsize=(8, 8))
nx.draw(G, node_size=30, with_labels=False, edge_color="gray")
plt.title("ErdsRnyi Random Graph G(100, 0.2, 0.1) with Self-loops")
plt.show()
```
