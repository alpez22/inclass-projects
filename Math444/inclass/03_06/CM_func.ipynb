```python
import numpy as np
import networkx as nx

def CM(d):
    # Configuration model with degree sequence d = [d1, ... , dn] (a list or array)
    # di \ge 0 and sum to even 
    d = list(d)
    stubs_list = []
    for i in np.arange(len(d)):
        for j in np.arange(d[i]):
            # d = [3,5] => stubs_list = [0,0], [0,1], [0,2], [1,0], [1,1], [1,2], [1,3], [1,4]
            stubs_list.append([i,j]) # nodes + half-edges
    #print(stubs_list)
    G = nx.MultiGraph()
    while len(stubs_list)>0:
        ss = np.random.choice(np.asarray(len(stubs_list)), 2, replace=False) #all half edges = len(stubs_list) = 8
        
        half_edge1 = stubs_list[ss[0]]
        half_edge2 = stubs_list[ss[1]]
        
        G.add_edge(half_edge1[0], half_edge2[1])
        stubs_list.remove(half_edge1)
        stubs_list.remove(half_edge2)
            
    return G
```

```python
d = [3] * 10
nx.draw(CM(d))
```
