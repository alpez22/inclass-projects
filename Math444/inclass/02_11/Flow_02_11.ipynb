```python
import networkx as nx # for handling graphs/networks 
import numpy as np # for basic scientific computing 
import matplotlib.pyplot as plt # for plotting

def create_example_network():
    G = nx.DiGraph()
    
    # Add edges with capacities
    edges = [
        ('s', 'a', 4), ('s', 'b', 2),
        ('a', 'b', 1), ('a', 'c', 2), ('a', 'd', 4),
        ('b', 'd', 2),
        ('c', 't', 3),
        ('d', 't', 3)
    ]
    
    G.add_weighted_edges_from(edges)
    return G

def plot_network(G, flow=None):
    pos = nx.spring_layout(G)
    plt.figure(figsize=(10, 6))
    
    # Draw the graph
    nx.draw_networkx_nodes(G, pos)
    
    # Draw edges with labels
    if flow:
        edge_labels = {(u, v): f'{flow[u][v]}/{d["weight"]}'
                      for (u, v, d) in G.edges(data=True)}
    else:
        edge_labels = {(u, v): f'{d["weight"]}'
                      for (u, v, d) in G.edges(data=True)}
    
    nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=15)
    nx.draw_networkx_edges(G, pos)
    nx.draw_networkx_labels(G, pos)
    
    plt.title('Flow Network')
    plt.axis('off')
    plt.show()

# Create and visualize the network
G = create_example_network()
plot_network(G)
```

```python
def find_path(G, source, sink, path, visited):
    ''' 
        Find aumenting path from source to sink
        Current flow is stored as edge weights in G
    '''
    if source==sink:
        return path

    visited.add(source)

    for neighbor in G.neighbors(source):
        if (neighbor not in visited and
            G.edges[source, neighbor]['weight'] > 0):
            result = find_path(G, neighbor, sink, path + [neighbor], visited)
            if result:
                return result
    return None

path = find_path(G, 's', 't', ['s'], set())
print("Augmenting path found =", path)
        
```

```python
def ford_fulkerson(G, source, sink):
    """Implement Ford-Fulkerson algorithm"""
    # Create residual graph
    residual = G.copy()
    
    # Initialize flow dictionary
    flow = {}
    for u in G.nodes():
        flow[u] = {}
        for v in G.nodes():
            flow[u][v] = 0
    
    # Find augmenting paths and update flow
    while True:
        path = find_path(residual, source, sink, [source], set())
        if not path:
            break
            
        # Find minimum residual capacity along the path
        path_flow = float('inf')
        for i in range(len(path)-1):
            u, v = path[i], path[i+1]
            path_flow = min(path_flow, residual.edges[u, v]['weight'])
        
        # Update residual capacities and flow
        for i in range(len(path)-1):
            u, v = path[i], path[i+1]
            flow[u][v] += path_flow
            flow[v][u] -= path_flow
            residual.edges[u, v]['weight'] -= path_flow
            
            # Add reverse edge if it doesn't exist
            if not residual.has_edge(v, u):
                residual.add_edge(v, u, weight=0)
            residual.edges[v, u]['weight'] += path_flow
    
    return flow

# Run Ford-Fulkerson on our example network
max_flow = ford_fulkerson(G, 's', 't')
plot_network(G, max_flow)
```

```python
def plot_step(G, flow, residual, pos, path=None, step_num=0):
    """Plot current flow and residual graphs side by side"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4))
    
    # Left plot remains the same
    ax1.set_title(f'Step {step_num}: Current Flow', fontsize=14)
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', ax=ax1)
    edge_labels = {(u, v): f'{flow[u][v]}/{d["weight"]}' 
                  for (u, v, d) in G.edges(data=True)}
    nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=12, ax=ax1)
    edges = list(G.edges())
    edge_colors = ['blue' if flow[u][v] > 0 else 'gray' for (u, v) in edges]
    nx.draw_networkx_edges(G, pos, edgelist=edges, edge_color=edge_colors, ax=ax1,
                          arrowsize=20)
    
    # Right plot (Residual Graph)
    ax2.set_title(f'Step {step_num}: Residual Graph', fontsize=14)
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', ax=ax2)
    
    # Get only edges with positive capacity
    forward_edges = [(u, v) for (u, v, d) in residual.edges(data=True) 
                    if d.get('type') == 'forward' and d['weight'] > 0]
    backward_edges = [(u, v) for (u, v, d) in residual.edges(data=True) 
                     if d.get('type') == 'backward' and d['weight'] > 0]
    
    def get_parallel_offset(pos_u, pos_v, offset=0.05):
        dx = pos_v[0] - pos_u[0]
        dy = pos_v[1] - pos_u[1]
        length = (dx**2 + dy**2)**0.5
        if length == 0:
            return 0, 0
        normalized_offset = (-dy/length * offset, dx/length * offset)
        return normalized_offset
    
    # Draw edges
    if path:
        path_edges = list(zip(path[:-1], path[1:]))
        # Only include path edges that still exist in residual graph and have positive capacity
        path_edges_with_capacity = [(u, v) for (u, v) in path_edges 
                                  if residual.has_edge(u, v) and 
                                  residual.edges[u, v]['weight'] > 0]
        
        # Draw path edges in red (only those with capacity)
        if path_edges_with_capacity:  # Only draw if there are valid edges
            nx.draw_networkx_edges(residual, pos, edgelist=path_edges_with_capacity, 
                                 edge_color='red', width=2, ax=ax2,
                                 arrowsize=20)
        
        other_forward = [e for e in forward_edges if e not in path_edges]
        other_backward = [e for e in backward_edges if e not in path_edges]
        
        # Draw non-path forward edges
        if other_forward:  # Only draw if there are edges
            nx.draw_networkx_edges(residual, pos, edgelist=other_forward,
                                 edge_color='gray', style='solid', ax=ax2,
                                 arrowsize=20)
        
        # Draw non-path backward edges with offset
        for u, v in other_backward:
            offset = get_parallel_offset(pos[u], pos[v])
            new_pos_u = (pos[v][0] + offset[0], pos[v][1] + offset[1])
            new_pos_v = (pos[u][0] + offset[0], pos[u][1] + offset[1])
            ax2.annotate("", xy=new_pos_u, xytext=new_pos_v,
                        arrowprops=dict(arrowstyle="->", color='gray', 
                                      linestyle='--', alpha=0.7,
                                      mutation_scale=20))
    else:
        # Draw forward edges
        if forward_edges:  # Only draw if there are edges
            nx.draw_networkx_edges(residual, pos, edgelist=forward_edges,
                                 edge_color='gray', style='solid', ax=ax2,
                                 arrowsize=20)
        
        # Draw backward edges with offset
        for u, v in backward_edges:
            offset = get_parallel_offset(pos[u], pos[v])
            new_pos_u = (pos[v][0] + offset[0], pos[v][1] + offset[1])
            new_pos_v = (pos[u][0] + offset[0], pos[u][1] + offset[1])
            ax2.annotate("", xy=new_pos_u, xytext=new_pos_v,
                        arrowprops=dict(arrowstyle="->", color='gray', 
                                      linestyle='--', alpha=0.7,
                                      mutation_scale=20))
    
    # Add edge labels only for edges with positive capacity
    for u, v, d in residual.edges(data=True):
        if d['weight'] > 0:  # Only label edges with positive capacity
            if d.get('type') == 'forward':
                edge_center = ((pos[u][0] + pos[v][0])/2, (pos[u][1] + pos[v][1])/2)
                ax2.annotate(str(d['weight']),
                            xy=edge_center,
                            xytext=(0, 5),
                            textcoords='offset points',
                            ha='center',
                            va='bottom',
                            bbox=dict(boxstyle='round,pad=0.2', fc='white', ec='none', alpha=0.7))
            else:
                offset = get_parallel_offset(pos[u], pos[v])
                edge_center = ((pos[u][0] + pos[v][0])/2 + offset[0],
                             (pos[u][1] + pos[v][1])/2 + offset[1])
                ax2.annotate(str(d['weight']),
                            xy=edge_center,
                            xytext=(0, 5),
                            textcoords='offset points',
                            ha='center',
                            va='bottom',
                            bbox=dict(boxstyle='round,pad=0.2', fc='white', ec='none', alpha=0.7))
    
    nx.draw_networkx_labels(G, pos, font_size=12, ax=ax2)
    
    plt.tight_layout()
    plt.show()

def ford_fulkerson_with_viz(G, source, sink):
    """Ford-Fulkerson with visualization of each step"""
    # Create residual graph with original edges and capacities
    residual = nx.DiGraph()
    
    # Initialize residual graph with forward edges only
    for u, v, data in G.edges(data=True):
        residual.add_edge(u, v, weight=data['weight'], type='forward')
    
    # Create and fix a layout
    pos = nx.spring_layout(G, seed=42)
    
    # Initialize flow dictionary
    flow = {}
    for u in G.nodes():
        flow[u] = {}
        for v in G.nodes():
            flow[u][v] = 0
    
    step = 0
    plot_step(G, flow, residual, pos=pos, step_num=step)
    
    while True:
        path = find_path(residual, source, sink, [source], set())
        if not path:
            break
            
        path_flow = float('inf')
        for i in range(len(path)-1):
            u, v = path[i], path[i+1]
            path_flow = min(path_flow, residual.edges[u, v]['weight'])
        
        print(f"Found augmenting path: {path} with flow {path_flow}")
        
        for i in range(len(path)-1):
            u, v = path[i], path[i+1]
            # Update flow
            flow[u][v] += path_flow
            flow[v][u] -= path_flow
            
            # Update forward edge
            forward_cap = residual.edges[u, v]['weight'] - path_flow
            if forward_cap > 0:
                residual.edges[u, v]['weight'] = forward_cap
            else:
                residual.remove_edge(u, v)
            
            # Add or update backward edge
            if residual.has_edge(v, u):
                residual.edges[v, u]['weight'] += path_flow
            else:
                residual.add_edge(v, u, weight=path_flow, type='backward')
            
            print(f"Edge {u}->{v}: Forward capacity now {forward_cap}, Backward capacity now {path_flow}")
        
        step += 1
        plot_step(G, flow, residual, path=path, pos=pos, step_num=step)
    
    return flow

# Create and run example
G = create_example_network()
max_flow = ford_fulkerson_with_viz(G, 's', 't')
```

```python
def find_min_cut(G, source, sink, flow):
    """Find the minimum cut in the network"""
    # Find set of nodes reachable from source in residual graph
    S = set([source])
    queue = [source]
    
    # Create residual graph
    residual = G.copy()
    for u, v, d in residual.edges(data=True):
        residual.edges[u, v]['weight'] -= flow[u][v]
    
    # BFS to find all reachable nodes
    while queue:
        u = queue.pop(0)
        for v in residual.neighbors(u):
            if v not in S and residual.edges[u, v]['weight'] > 0:
                S.add(v)
                queue.append(v)
    
    # Find cut edges
    cut_edges = [(u, v) for u, v, d in G.edges(data=True)
                 if u in S and v not in S]
    
    return S, cut_edges

def plot_min_cut(G, S, cut_edges):
    pos = nx.spring_layout(G)
    plt.figure(figsize=(10, 6))
    
    # Draw nodes with different colors for S and T
    nx.draw_networkx_nodes(G, pos, nodelist=S, node_color='lightblue')
    nx.draw_networkx_nodes(G, pos, nodelist=[n for n in G.nodes() if n not in S],
                          node_color='lightgreen')
    
    # Draw cut edges in red, other edges in black
    nx.draw_networkx_edges(G, pos, edgelist=cut_edges, edge_color='red', width=2)
    other_edges = [(u, v) for (u, v) in G.edges() if (u, v) not in cut_edges]
    nx.draw_networkx_edges(G, pos, edgelist=other_edges)
    
    # Draw labels
    nx.draw_networkx_labels(G, pos)
    
    plt.title('Minimum Cut')
    plt.axis('off')
    plt.show()

# Find and visualize the minimum cut
S, cut_edges = find_min_cut(G, 's', 't', max_flow)
plot_min_cut(G, S, cut_edges)
```
